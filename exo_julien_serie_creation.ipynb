Une série linéaire y(t) = a*t+b

```python
def func(a,b):
    print ("F(x) = ax + b")
    x = np.arange(100)
    Fx = a * x + b
    return Fx
```

```python
serie_2 = func(2,3)
plt.plot(serie_2)
plt.show
```

une série oscillante y(t) = A * sin(omega*t+fi)

```python
def osc(A, omega, fi):
    print ("y(x) = A*sin(omega*x + fi")
    x = np.arange(100)
    yx = A * np.sin(omega*x + fi)
    return yx
```

```python
serie_3 = osc(2,2,2)
plt.plot(serie_3)
plt.show
```
