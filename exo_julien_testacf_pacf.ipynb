```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Create a new time serie

```python
serie_1 = np.arange(1000)
noise = np.random.normal(loc=0.0, scale=1.0, size=1000)
plt.plot(noise)
plt.show
```

# Plot ACF et PACF

```python
from statsmodels.graphics.tsaplots import plot_acf

# the autocorrelation chart provides just the correlation at increasing lags
fig, ax = plt.subplots(figsize=(12,5))
plot_acf(noise, lags=20, ax=ax)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf

fig, ax = plt.subplots(figsize=(12,5))
plot_pacf(noise, lags=20, ax=ax)
plt.show()
```
